# Overview


There are 4 main functions of EZFacialRecognition:


```
checkCameras(numberOfCameras)
giveOption()
showVid(drawLines=False)
detect(option, timeout=None)
```


<h3>Check Cameras

works by cycling through an indexed list of cameras in order to the total number of cameras that are operational. ie. If you hvae 3 cameras and 2 are active, then 1 is offline.

<h3>Gve Option

This function dispalys the options that come with the program. This is what runs when EZFacialRecognition.py is run by itself rather than imported.

<h3> Show Vid

This function displays the output of the first camera and has a parameter to draw the facial detect lines or not

<h3>Detect

This function will pause the program until a face is detected, then it returns true it has a timeout and an option parameter. The option parameter allows the user to select what is being detected ie. eyes, face, body.
